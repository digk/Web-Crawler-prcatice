```python
#!/usr/bin/python
# coding: utf-8
import requests #引入包request
from bs4 import BeautifulSoup #从bs4这个库中导入BeautifulSoup
link="http://www.santostang.com/" # 定义link为目标网页地址
# 定义请求头的浏览器代理，伪装成浏览器
headers = {'User-Agent':'Mozilla/5.0(Windows;U;Windows NT 6.1;en-US;rv:1.9.1.6)Gecko/20091201 Firefox/3.5.6'}
r = requests.get(link,headers=headers)#请求网页
soup = BeautifulSoup(r.text,"html.parser") #使用BeautifulSoup 解析  
# 找到第一篇文章标题，定位到class是“post-title”的h1元素，提取a，提取a里面的字符串，strip（）去除左右空格
title = soup.find("h1", class_ = "post-title").a.text.strip()
# print(r.text) #r.text 是获取网页的页面内容
print(title)
```

```python
import requests #引入包request
from bs4 import BeautifulSoup #从bs4这个库中导入BeautifulSoup
link="http://www.santostang.com/" # 定义link为目标网页地址
# 定义请求头的浏览器代理，伪装成浏览器
headers = {'User-Agent':'Mozilla/5.0(Windows;U;Windows NT 6.1;en-US;rv:1.9.1.6)Gecko/20091201 Firefox/3.5.6'}
r = requests.get(link,headers=headers)#请求网页
soup = BeautifulSoup(r.text,"html.parser") #使用BeautifulSoup 解析  
# 找到第一篇文章标题，定位到class是“post-title”的h1元素，提取a，提取a里面的字符串，strip（）去除左右空格
title = soup.find("h1", class_ = "post-title").a.text.strip()
# print(r.text) #r.text 是获取网页的页面内容
print(title)
# 打开一个空白的TXT，然后使用f.write写入刚刚的字符串title
with open('title_test.txt', "a+") as f:
    f.write(title)
```

```python
import requests
r = requests.get('http://www.santostang.com/')
print("文本编码：", r.encoding)   # r.encoding 是服务器内容使用的文本编码
print("响应状态码：", r.status_code)  #  r.status_code 用于检测响应的状态码，如果返回200，就表示请示成功了；如果返回的是4XX，就表示客户端错误；返回5XX则表示服务器错误响应。
                                    #  我们可以用r.status_code 来检测请求是否正确响应
print("字符串方式的响应体：", r.text)  #r.text是服务器响应的内容，会自动根据响应头部的字符串编码进行解码。
```

（1） r.text是服务器响应的内容，会自动根据响应头部的字符串编码进行解码。
（2） r.encoding 是服务器内容使用的文本编码。
（3） r.status_code 用于检测响应的状态码，如果返回200，就表示请示成功了；如果返回的是4XX，就表示客户端错误；返回5XX则表示服务器错误响应。我们可以用r.status_code 来检测请求是否正确响应。
（4） r.content 是字节方式的响应体，会自动解码gzip和deflate编码的响应数据。
（5） r.json() 是Request中内置的JSON解码器。

```python
import requests
key_dict = {'key1': 'value1', 'key2': 'value2'}
r = requests.get('http://httpbin.org/get', params = key_dict)
print("URL已经正确解码：", r.url)
print("字符串方式的响应体：\n", r.text)
```
